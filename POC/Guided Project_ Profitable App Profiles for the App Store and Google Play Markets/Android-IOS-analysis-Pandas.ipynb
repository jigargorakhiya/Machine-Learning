```python
def is_english(string):
    non_ascii = 0
    for char in string:
        if ord(char) > 127:
            non_ascii +=1
    
    if non_ascii > 3:
        return False
    else:
        return True
# print(is_english('Docs To Go™ Free Office Suite'))
# print(is_english('Instachat 😜'))
```

```python
android_clean['isEng'] = android_clean.App.apply(is_english)
ios['isEng'] = ios.track_name.apply(is_english)
```

## Filter Free Apps

```python
def isFreeApp(price):
    try:
        price = float(price)
    except:
        price = float(price[1:])
    if price == 0.0:
        return 0 # Free
    else:
        return 1 # Paid
```

```python
android_eng['isFree'] = android_eng.Price.apply(isFreeApp)
android_free = android_eng[android_eng.isFree == 0] ## Free App

ios_eng['isFree'] = ios_eng.price.apply(isFreeApp)
ios_free = ios_eng[ios_eng.isFree == 0] ## Free App
```

## Most common apps by Genere

```python
def freq_table(count_genere,length_data):
    return float( round((count_genere/length_data)*100 ,2 ))
```

```python
def groupByCol(df,groupby,count_col):
    df_cat = df.groupby(by=groupby)[count_col].count()
    df_cat = pd.DataFrame(df_cat).reset_index()
    df_cat.columns = [groupby, 'col_count']
    
    df_cat['percentage'] = df_cat.col_count.apply(freq_table,length_data=len(df))
    df_cat.sort_values(by='percentage',ascending=False,inplace=True)
    
    return df_cat
```

## Avg user_Rating for ios

```python
def ios_user_rating_table1(dataset,index):
    data_genere={}
    user_ratings={}
    
    for data in dataset:
        genere = data[index]
        if genere in user_ratings:
            user_ratings[genere] += float(data[5])
            data_genere[genere] += 1 
        else:
            user_ratings[genere] = float(data[5])
            data_genere[genere] = 1 
    
    freq = {}       
    for key,value in user_ratings.items():
        freq[key] = round(value/data_genere[key],2)
    return freq

#ios_user_rating_table1(ios_free,11)
```
