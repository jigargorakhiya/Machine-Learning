```python
import matplotlib.pyplot as plt

%matplotlib inline #Do not need to wirte plt.show because of this line
```

```python
import numpy as np
```

```python
x = np.arange(0,10)
y= np.arange(11,21)
```

```python
#Plotting using mapplot

plt.scatter(x,y,c="r")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("2D Graph")

plt.savefig("Graph.png")
```

```python
#Plt

y = x *x
```

```python
plt.plot(x,y,'r-*',linewidth=1,markersize=12)
```

# Creating Subplot

```python
plt.subplot(2,3,1)
plt.plot(x,y,'g+')

plt.subplot(2,3,2)
plt.plot(x,y,'r*--')

plt.subplot(2,3,3)
plt.plot(x,y,'g*--')

plt.subplot(2,3,4)
plt.plot(x,y,'b+--')
```

```python
#Sine Curv

x = np.arange(0, 2 * np.pi, 0.1)
y= np.sin(x)

print(x)
print(y)

plt.plot(x,y)
```

```python
# Sin & Cos Curve
```

```python
x = np.arange(0, 2 * np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2,1,1)
plt.plot(x,y_sin,'r--')
plt.title("Sine Curv")

plt.subplot(2,1,2)
plt.plot(x,y_cos,'g--')
plt.title("Cosine Curv")
```

```python
#Bar Graph

x = [2,4,10]
y = [5,10,20]

x1 = [3,6,9]
y1 = [9,15,12]

plt.bar(x,y,align="center")
plt.bar(x1,y1)
plt.title("Bar Graph")
```
