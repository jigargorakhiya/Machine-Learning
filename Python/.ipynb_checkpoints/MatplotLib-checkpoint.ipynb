```python
import matplotlib.pyplot as plt

%matplotlib inline #Do not need to wirte plt.show because of this line
```

```python
import numpy as np
```

```python
x = np.arange(0,10)
y= np.arange(11,21)
```

```python
#Plotting using mapplot

plt.scatter(x,y,c="r")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("2D Graph")

plt.savefig("Graph.png")
```

```python
#Plt

y = x *x
```

```python
plt.plot(x,y,'r-*',linewidth=1,markersize=12)
```

# Creating Subplot

```python
plt.subplot(2,3,1)
plt.plot(x,y,'g+')

plt.subplot(2,3,2)
plt.plot(x,y,'r+--')

plt.subplot(2,3,3)
plt.plot(x,y,'g+--')

plt.subplot(2,3,2)
plt.plot(x,y,'g+--')
```
